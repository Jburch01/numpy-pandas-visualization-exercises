```python
from matplotlib import pyplot as plt
import pandas as pd
from pydataset import data
import numpy as np
import random
import math
```

```python
n = np.array([i + 1 for i  in range(10)])
```

```python
y1 = 0 * n + 1
y2 = np.log(n)
y3 = n
y4 = n * y2
y5 = n ** 2
y6 = 2 ** n
y7 = [np.math.factorial(n) for n in n]
y8 = n ** n
```

```python
fig, ax = plt.subplots()

plt.xlabel('Elements')
plt.ylabel("Operations")
plt.title("Big O Notation")

ax.plot(n, y1, label="O(1)")
ax.plot(n, y2, label="O(log (n))")
ax.plot(n, y3, label="O(n)")
ax.plot(n, y4, label="O(n log n)")
ax.plot(n, y5, label="O(n^2)")
ax.plot(n, y6, label="O(2^n)")
ax.plot(n, y7, label="O(n!)")
ax.plot(n, y8, label="n^n")

plt.xlim(1, 10)
plt.ylim(0, 100)


ax.legend(loc='upper right', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)

plt.show()
```
