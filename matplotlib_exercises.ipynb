#### 1. Use matplotlib to plot the following equation: y = x**2 - x + 2



```python
x = np.array([i for i in range(10)])
y = x**2 - x + 2
```

```python
plt.plot(x, y, color="purple")

plt.xlabel('X')
plt.ylabel("Y")
plt.title("y = x **2 -x + 2")
plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'orange'})
plt.show()

```
